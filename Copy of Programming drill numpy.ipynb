```
# Indexing
import numpy as np
a = np.identity(5)
print('5th row in the 2-d array is:', a[4,:]) #indices begins at 0
a[2,4] = 100; #assigning value in the matrix
print(a)

print('-'*70)

# Slicing
b = np.arange(20)
print(b[1:18:2]) #[starindex: endindex(exclusive) : step]

print('-'*70)

#Basic Operations
a = np.array([1,2,3,4])
print(a+1) # This adds 1 to each element in array a
print(a**2) # squares each element in array a

print('-'*70)

#Matrix Multiplication
c = np.diag([1,2,3,4])
d = np.ones(4) +1
print(c*d)
```

```
import numpy as np
a1 = np.arange(1,7).reshape(2,3)##Array woth numbers 1,2,...,6
a2 = np.arange(1,7).reshape(2,3)*10 ##Array with numbers 10, 20,,...,60

print('a1: \n',a1, '\n a2: \n' , a2)

print('-'*70)

##Now let us stack these arrays horizontally, one to the right and another to the left

a_horizontally_stacked = np.hstack([a1,a1])
print('Horizontal stacking: \n', a_horizontally_stacked)
print('Shape of array after horizonatal stacking:',a_horizontally_stacked.shape)

print('-'*70)

##Now let us stack these arrays vertically, a1 on top and a2 at the bottom
##So shape of arrays after stacking will be [4,3]
##Always be mindful of shape of the arrays. This will help you in debugging.
a_vertically_stacked = np.vstack([a1, a2])

print('Vertical stacking: \n',a_vertically_stacked)
print('Shape of array after vertical stacking:', a_vertically_stacked.shape)
```
