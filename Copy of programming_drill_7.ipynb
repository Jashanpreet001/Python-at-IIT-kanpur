```
# indexing 
import numpy as np
a = np.identity(5)
print('5th row in the 2-d array is:', a[4, :]) #indexing begin at 0
a[2, 4] = 100; #assiging value in the matrix
print(a)

print('-'*70)

#slicing
b = np.arange(20)
print(b[1:18:2]) #[startindex: endindex(exclusive) : step]

print('-'*70)

#Basic Operations
a = np.array([1, 2, 3, 4])
print(a + 1) #This adds 1 to each element in array a
print(a**2) #squares each element in array a

print('-'*70)

#Matrix Multiplication
c = np.diag([1, 2, 3, 4])
d = np.ones(4) + 1
print(c * d)
```
