##Q. Interest Compounded
The Formula for computing compound interest is as follows - 
\begin{equation}
A = P(1 + \frac{r}{n})^{nt}
\end{equation}
Write a program that takes as inputs $P, r, n, t$ and prints the total amount that has to be paid.

```
P= int(input("Enter the principle amount:\n"))
r= int(input("Enter the rate of interest:\n"))
n= int(input("Enter the number of times interest applied per year:\n"))
t= int(input("Enter the number of years:\n"))

compound_interest = P*((1+(r/n))**(n*t))
print("The final amount is",compound_interest)
```
